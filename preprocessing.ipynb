```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\jetha\Downloads\archive\Plant_1_Generation_Data.csv")
df.head()
```

```python
df2=pd.read_csv(r"C:\Users\jetha\Downloads\archive\Plant_2_Generation_Data.csv")
df2.head()
```

```python
#combine both dataframes
df3=pd.concat([df,df2])
df3.head()
```

```python
#make a new CSV file
df3.to_csv('Combined_Plant_Generation_Data.csv', index=False)
```

```python
#form a pie chart for plant ID 4135001 and 4136001 for the total yield
df3['TOTAL_YIELD']=df3['DAILY_YIELD']+df3['TOTAL_YIELD']
df3['TOTAL_YIELD'].head()
df3['PLANT_ID'].head()
df3['PLANT_ID'].value_counts()
df3['PLANT_ID'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Plant ID 4135001 and 4136001')
plt.show()
```
