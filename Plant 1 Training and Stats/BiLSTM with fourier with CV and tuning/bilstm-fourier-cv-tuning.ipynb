```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import sin, cos, pi
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import optuna
from optuna.samplers import TPESampler
import warnings
warnings.filterwarnings('ignore')

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class SolarEnergyForecaster:
    def __init__(self, plant_file='/kaggle/input/solar-power-generation-data/Plant1_filtered.csv',
                 weather_file='/kaggle/input/solar-power-generation-data/Plant1_Weather_filtered.csv'):
        self.plant_file = plant_file
        self.weather_file = weather_file
        self.scaler_X = MinMaxScaler()
        self.scaler_y = MinMaxScaler()
        self.best_models = {}
        self.cv_results = {}

    def load_and_preprocess_data(self):
        """Load and preprocess the plant and weather data"""
        print("Loading and preprocessing data...")

        # Load data
        plant_df = pd.read_csv(self.plant_file)
        weather_df = pd.read_csv(self.weather_file)

        # Convert datetime
        plant_df['DATE_TIME'] = pd.to_datetime(plant_df['DATE_TIME'])
        weather_df['DATE_TIME'] = pd.to_datetime(weather_df['DATE_TIME'])

        # Aggregate plant data (20 source keys per timestamp → mean AC_POWER)
        plant_agg = plant_df.groupby('DATE_TIME').agg({'AC_POWER':'mean'}).reset_index()

        # Merge datasets
        df = pd.merge(weather_df, plant_agg, on='DATE_TIME', how='inner')
        df.set_index('DATE_TIME', inplace=True)

        # Create Fourier features for temporal patterns
        df['hour'] = df.index.hour + df.index.minute/60.0
        df['day_of_year'] = df.index.dayofyear
        df['sin_hour'] = np.sin(2*pi*df['hour']/24)
        df['cos_hour'] = np.cos(2*pi*df['hour']/24)
        df['sin_doy'] = np.sin(2*pi*df['day_of_year']/365)
        df['cos_doy'] = np.cos(2*pi*df['day_of_year']/365)
        df.drop(['hour','day_of_year'], axis=1, inplace=True)

        # Select numeric features
        target_col = 'AC_POWER'
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        feature_cols = [c for c in numeric_cols if c != target_col]

        print(f"Feature columns: {feature_cols}")
        print(f"Data shape: {df.shape}")

        self.df = df
        self.target_col = target_col
        self.feature_cols = feature_cols

        return df

    def create_sequences(self, X, y, time_steps, horizon):
        """Create sequences for LSTM training"""
        Xs, ys = [], []
        for i in range(len(X) - time_steps - horizon + 1):
            Xs.append(X[i : i+time_steps])
            ys.append(y[i+time_steps : i+time_steps+horizon].flatten())
        return np.array(Xs), np.array(ys)

    def build_model(self, input_shape, horizon, lstm1_units=128, lstm2_units=64,
                    dropout_rate=0.3, learning_rate=0.001, batch_size=32):
        """Build BiLSTM model with given hyperparameters"""
        model = Sequential([
            Bidirectional(LSTM(lstm1_units, return_sequences=True), input_shape=input_shape),
            Dropout(dropout_rate),
            Bidirectional(LSTM(lstm2_units)),
            Dropout(dropout_rate),
            Dense(horizon)
        ])

        optimizer = Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
        return model

    def time_series_cv_score(self, X, y, horizon, hyperparams, n_splits=5):
        """Perform time series cross-validation"""
        tscv = TimeSeriesSplit(n_splits=n_splits)
        cv_scores = []

        for fold, (train_idx, val_idx) in enumerate(tscv.split(X)):
            X_train_fold = X[train_idx]
            y_train_fold = y[train_idx]
            X_val_fold = X[val_idx]
            y_val_fold = y[val_idx]

            # Build and train model
            model = self.build_model(
                input_shape=(X.shape[1], X.shape[2]),
                horizon=horizon,
                **hyperparams
            )

            # Callbacks
            es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0)
            rl = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=0)

            # Train with validation split from training data
            history = model.fit(
                X_train_fold, y_train_fold,
                validation_split=0.1,
                epochs=50,  # Reduced for faster HPT
                batch_size=hyperparams.get('batch_size', 32),
                callbacks=[es, rl],
                verbose=0
            )

            # Predict and calculate metrics
            y_pred = model.predict(X_val_fold, verbose=0)

            # Calculate RMSE for the fold
            rmse = np.sqrt(mean_squared_error(y_val_fold.flatten(), y_pred.flatten()))
            cv_scores.append(rmse)

            # Clean up
            del model
            tf.keras.backend.clear_session()

        return np.mean(cv_scores)

    def objective(self, trial, X, y, horizon):
        """Optuna objective function for hyperparameter optimization"""
        # Define hyperparameter search space
        hyperparams = {
            'lstm1_units': trial.suggest_categorical('lstm1_units', [64, 128, 256]),
            'lstm2_units': trial.suggest_categorical('lstm2_units', [32, 64, 128]),
            'dropout_rate': trial.suggest_float('dropout_rate', 0.1, 0.5),
            'learning_rate': trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True),
            'batch_size': trial.suggest_categorical('batch_size', [16, 32, 64])
        }

        # Perform cross-validation
        cv_score = self.time_series_cv_score(X, y, horizon, hyperparams, n_splits=3)
        return cv_score

    def hyperparameter_tuning(self, X, y, horizon, n_trials=50):
        """Perform hyperparameter tuning using Optuna"""
        print(f"Starting hyperparameter tuning for horizon {horizon} with {n_trials} trials...")

        # Create study
        study = optuna.create_study(
            direction='minimize',
            sampler=TPESampler(seed=42),
            study_name=f'solar_forecast_h{horizon}'
        )

        # Optimize
        study.optimize(lambda trial: self.objective(trial, X, y, horizon), n_trials=n_trials)

        print(f"Best trial for horizon {horizon}:")
        print(f"  Value: {study.best_trial.value:.4f}")
        print(f"  Params: {study.best_trial.params}")

        return study.best_trial.params, study.best_trial.value

    def train_final_model(self, X_train, y_train, X_test, y_test, horizon, best_params):
        """Train final model with best hyperparameters"""
        print(f"Training final model for horizon {horizon}...")

        # Build model with best parameters
        model = self.build_model(
            input_shape=(X_train.shape[1], X_train.shape[2]),
            horizon=horizon,
            **best_params
        )

        # Callbacks
        es = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)
        rl = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=7, min_lr=1e-6, verbose=1)

        # Train model
        history = model.fit(
            X_train, y_train,
            validation_split=0.1,
            epochs=100,
            batch_size=best_params.get('batch_size', 32),
            callbacks=[es, rl],
            verbose=1
        )

        return model, history

    def evaluate_model(self, model, X_test, y_test, horizon):
        """Evaluate model performance"""
        # Make predictions
        y_pred = model.predict(X_test, verbose=0)

        # Inverse transform if needed
        def inverse_transform_if_needed(y_scaled):
            if hasattr(self, 'scaler_y'):
                # Reshape y_scaled to be 2D before inverse_transform
                flat = y_scaled.reshape(-1, 1)
                inverted = self.scaler_y.inverse_transform(flat)
                # Reshape back to original shape but without the extra dimension if it exists
                return inverted.reshape(y_scaled.shape)
            return y_scaled

        y_test_inv = inverse_transform_if_needed(y_test)
        y_pred_inv = inverse_transform_if_needed(y_pred)

        # Calculate metrics
        # Ensure arrays are flattened or reshaped to 1D for mean_absolute_error and mean_squared_error
        mae = mean_absolute_error(y_test_inv.flatten(), y_pred_inv.flatten())
        rmse = np.sqrt(mean_squared_error(y_test_inv.flatten(), y_pred_inv.flatten()))
        r2 = r2_score(y_test_inv.flatten(), y_pred_inv.flatten())

        # First-step metrics
        mae_first = mean_absolute_error(y_test_inv[:, 0].flatten(), y_pred_inv[:, 0].flatten())
        rmse_first = np.sqrt(mean_squared_error(y_test_inv[:, 0].flatten(), y_pred_inv[:, 0].flatten()))
        r2_first = r2_score(y_test_inv[:, 0].flatten(), y_pred_inv[:, 0].flatten())

        metrics = {
            'mae_all': mae,
            'rmse_all': rmse,
            'r2_all': r2,
            'mae_first': mae_first,
            'rmse_first': rmse_first,
            'r2_first': r2_first
        }

        return metrics, y_pred_inv, y_test_inv

    def plot_results(self, history, y_test, y_pred, horizon, metrics):
        """Plot training history and predictions"""
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))

        # Plot training history
        axes[0].plot(history.history['loss'], label='Train Loss')
        axes[0].plot(history.history['val_loss'], label='Validation Loss')
        axes[0].set_title(f'Training History (Horizon={horizon})')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('MSE Loss')
        axes[0].legend()
        axes[0].grid(True)

        # Plot predictions vs actual (first step)
        n_samples = min(200, len(y_test))
        axes[1].plot(y_test[:n_samples, 0], label='Actual', alpha=0.8)
        axes[1].plot(y_pred[:n_samples, 0], label='Predicted', alpha=0.8)
        axes[1].set_title(f'Predictions vs Actual (Horizon={horizon}, First Step)')
        axes[1].set_xlabel('Sample')
        axes[1].set_ylabel('AC_POWER')
        axes[1].legend()
        axes[1].grid(True)

        plt.tight_layout()
        plt.show()

        # Print metrics
        print(f"\nMetrics for Horizon {horizon}:")
        print(f"  First Step - MAE: {metrics['mae_first']:.2f}, RMSE: {metrics['rmse_first']:.2f}, R²: {metrics['r2_first']:.4f}")
        print(f"  All Steps  - MAE: {metrics['mae_all']:.2f}, RMSE: {metrics['rmse_all']:.2f}, R²: {metrics['r2_all']:.4f}")

    def run_experiment(self, time_steps=24, horizons=[1, 5, 24, 72], test_frac=0.2, n_trials=30):
        """Run the complete experiment with CV and HPT"""
        print("=== Solar Energy Forecasting with BiLSTM, CV, and HPT ===\n")

        # Load and preprocess data
        df = self.load_and_preprocess_data()

        # Scale features and target
        X_all = self.scaler_X.fit_transform(df[self.feature_cols])
        y_all = self.scaler_y.fit_transform(df[[self.target_col]])

        print(f"Scaled data shapes: X={X_all.shape}, y={y_all.shape}")

        results = {}

        for horizon in horizons:
            print(f"\n{'='*50}")
            print(f"PROCESSING HORIZON = {horizon}")
            print(f"{'='*50}")

            # Create sequences
            X, y = self.create_sequences(X_all, y_all, time_steps, horizon)
            print(f"Sequence shapes: X={X.shape}, y={y.shape}")

            # Split data
            split_idx = int(len(X) * (1 - test_frac))
            X_train, X_test = X[:split_idx], X[split_idx:]
            y_train, y_test = y[:split_idx], y[split_idx:]

            print(f"Train shapes: X={X_train.shape}, y={y_train.shape}")
            print(f"Test shapes: X={X_test.shape}, y={y_test.shape}")

            # Hyperparameter tuning
            best_params, best_score = self.hyperparameter_tuning(X_train, y_train, horizon, n_trials)

            # Train final model
            model, history = self.train_final_model(X_train, y_train, X_test, y_test, horizon, best_params)

            # Evaluate model
            metrics, y_pred, y_test_inv = self.evaluate_model(model, X_test, y_test, horizon)

            # Store results
            results[horizon] = {
                'best_params': best_params,
                'best_cv_score': best_score,
                'model': model,
                'history': history,
                'metrics': metrics,
                'y_pred': y_pred,
                'y_test': y_test_inv
            }

            # Plot results
            self.plot_results(history, y_test_inv, y_pred, horizon, metrics)

        self.results = results

        # Print summary
        self.print_summary()

        return results

    def print_summary(self):
        """Print experiment summary"""
        print(f"\n{'='*60}")
        print("EXPERIMENT SUMMARY")
        print(f"{'='*60}")

        summary_data = []
        for horizon, result in self.results.items():
            metrics = result['metrics']
            summary_data.append({
                'Horizon': horizon,
                'CV_Score': result['best_cv_score'],
                'Test_MAE': metrics['mae_first'],
                'Test_RMSE': metrics['rmse_first'],
                'Test_R²': metrics['r2_first'],
                'Best_Params': str(result['best_params'])
            })

        summary_df = pd.DataFrame(summary_data)
        print(summary_df.to_string(index=False))

        # Best performing horizon
        best_horizon = min(self.results.keys(),
                          key=lambda h: self.results[h]['metrics']['rmse_first'])
        print(f"\nBest performing horizon (lowest RMSE): {best_horizon}")
        print(f"Best RMSE: {self.results[best_horizon]['metrics']['rmse_first']:.2f}")

# Usage example
if __name__ == "__main__":
    # Initialize forecaster
    forecaster = SolarEnergyForecaster()

    # Run experiment
    results = forecaster.run_experiment(
        time_steps=24,
        horizons=[1, 5, 24, 72],
        test_frac=0.2,
        n_trials=30  # Adjust based on your computational resources
    )

    print("\nExperiment completed successfully!")
```
