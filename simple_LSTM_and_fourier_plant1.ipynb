# LSTM on Plant 1

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import datetime

# Load and preprocess generation data
gen_data = pd.read_csv('Datasets\Processed Datasets\Plant1_filtered.csv')  # Adjust path
gen_data['DATE_TIME'] = pd.to_datetime(gen_data['DATE_TIME'], format='%d-%m-%Y %H:%M')

# Aggregate AC_POWER across inverters
gen_agg = gen_data.groupby('DATE_TIME')['AC_POWER'].sum().reset_index()
gen_agg.columns = ['DATE_TIME', 'total_AC_POWER']

# Load and preprocess weather data
weather_data = pd.read_csv('Datasets\Processed Datasets\Plant1_Weather_filtered.csv')  # Adjust path
weather_data['DATE_TIME'] = pd.to_datetime(weather_data['DATE_TIME'])

# Merge datasets
data = pd.merge(gen_agg, weather_data[['DATE_TIME', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']], 
                on='DATE_TIME', how='inner')

# Add time features
data['hour'] = data['DATE_TIME'].dt.hour
data['minute'] = data['DATE_TIME'].dt.minute
data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)

# Define features and target
features = ['total_AC_POWER', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION', 'hour_sin', 'hour_cos']
target = 'total_AC_POWER'

# Normalize features
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[features])
scaled_df = pd.DataFrame(scaled_data, columns=features, index=data.index)
scaled_df['DATE_TIME'] = data['DATE_TIME']

# Define horizons
horizons = [4, 20, 96, 288]  # 1h, 5h, 1d, 3d (in 15-min intervals)
horizons_labels = ['1 hour', '5 hours', '1 day', '3 days']
max_horizon = max(horizons)

# Create sequences for each horizon
def create_sequences_single_horizon(df, seq_length, horizon):
    X, y = [], []
    for i in range(len(df) - seq_length - horizon):
        X.append(df[features].iloc[i:i+seq_length].values)
        # Target is the average AC_POWER over the horizon
        y_avg = df[target].iloc[i+seq_length:i+seq_length+horizon].mean()
        y.append(y_avg)
    return np.array(X), np.array(y)

seq_length = 4  # 1 hour (4 * 15 min)

# Train separate models for each horizon
models = {}
histories = {}
predictions = {}
test_actuals = {}

for horizon_idx, horizon in enumerate(horizons):
    print(f"\nTraining model for {horizons_labels[horizon_idx]} horizon...")
    
    # Create sequences for current horizon
    X, y = create_sequences_single_horizon(scaled_df, seq_length, horizon)
    
    # Split data chronologically
    train_size = int(0.8 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    # Build LSTM model for this horizon
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(seq_length, len(features)), return_sequences=False))
    model.add(Dense(1))  # Single output for this horizon
    model.compile(optimizer='adam', loss='mse')
    
    # Train model
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1, verbose=1)
    
    # Store model and history
    models[horizons_labels[horizon_idx]] = model
    histories[horizons_labels[horizon_idx]] = history
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Inverse scale predictions (target is first column in scaler)
    min_ac = scaler.data_min_[0]
    scale_ac = scaler.scale_[0]
    y_test_rescaled = y_test / scale_ac + min_ac
    y_pred_rescaled = y_pred.flatten() / scale_ac + min_ac
    
    # Store predictions and actuals
    predictions[horizons_labels[horizon_idx]] = y_pred_rescaled
    test_actuals[horizons_labels[horizon_idx]] = y_test_rescaled
    
    # Evaluate
    mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
    rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
    print(f'{horizons_labels[horizon_idx]} - MAE: {mae:.2f}, RMSE: {rmse:.2f}')
    
    # Save model
    model.save(f'plant1_lstm_model_{horizons_labels[horizon_idx].replace(" ", "_")}.h5')

# Baseline: Persistence model for each horizon
print("\nBaseline Persistence Model Results:")
for horizon_idx, horizon in enumerate(horizons):
    X, y = create_sequences_single_horizon(scaled_df, seq_length, horizon)
    train_size = int(0.8 * len(X))
    y_test = y[train_size:]
    
    # Persistence: use last value of sequence as prediction
    persistence_pred = X[train_size:, -1, 0]  # Last AC_POWER value from each sequence
    
    # Inverse scale
    min_ac = scaler.data_min_[0]
    scale_ac = scaler.scale_[0]
    y_test_rescaled = y_test / scale_ac + min_ac
    persistence_pred_rescaled = persistence_pred / scale_ac + min_ac
    
    persistence_mae = mean_absolute_error(y_test_rescaled, persistence_pred_rescaled)
    persistence_rmse = np.sqrt(mean_squared_error(y_test_rescaled, persistence_pred_rescaled))
    print(f'{horizons_labels[horizon_idx]} - Persistence MAE: {persistence_mae:.2f}, RMSE: {persistence_rmse:.2f}')

# Plot results for all horizons
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for i, horizon_label in enumerate(horizons_labels):
    axes[i].plot(test_actuals[horizon_label][:100], label='Actual', alpha=0.7)
    axes[i].plot(predictions[horizon_label][:100], label='Predicted', alpha=0.7)
    axes[i].set_title(f'{horizon_label} Average Prediction')
    axes[i].set_xlabel('Time Step')
    axes[i].set_ylabel('AC_POWER (kW)')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Inference: Predict future averages for all horizons
print("\nEnter current data for prediction:")
current_ac = float(input("Enter current total_AC_POWER (0 to 10000): "))
current_amb_temp = float(input("Enter current AMBIENT_TEMPERATURE (0 to 50): "))
current_mod_temp = float(input("Enter current MODULE_TEMPERATURE (0 to 70): "))
current_irr = float(input("Enter current IRRADIATION (0 to 1.5): "))

# Create sequence by repeating current input
current_time = datetime.datetime.now()
sequence_times = [current_time - datetime.timedelta(minutes=15*(seq_length - 1 - i)) for i in range(seq_length)]
sequence_df = pd.DataFrame({
    'DATE_TIME': sequence_times,
    'total_AC_POWER': [current_ac] * seq_length,
    'AMBIENT_TEMPERATURE': [current_amb_temp] * seq_length,
    'MODULE_TEMPERATURE': [current_mod_temp] * seq_length,
    'IRRADIATION': [current_irr] * seq_length,
})
sequence_df['hour'] = sequence_df['DATE_TIME'].dt.hour
sequence_df['minute'] = sequence_df['DATE_TIME'].dt.minute
sequence_df['hour_sin'] = np.sin(2 * np.pi * sequence_df['hour'] / 24)
sequence_df['hour_cos'] = np.cos(2 * np.pi * sequence_df['hour'] / 24)
sequence_features = sequence_df[features]
sequence_scaled = scaler.transform(sequence_features)

# Predict for all horizons using respective models
min_ac = scaler.data_min_[0]
scale_ac = scaler.scale_[0]

print("\nPredictions for all horizons:")
for horizon_label in horizons_labels:
    prediction_scaled = models[horizon_label].predict(sequence_scaled[np.newaxis, :, :])[0, 0]
    prediction_rescaled = prediction_scaled / scale_ac + min_ac
    print(f"Predicted average AC_POWER for next {horizon_label}: {prediction_rescaled:.2f} kW")
```

# Improvements on Plant 1 LSTM with Fourier Approximation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.layers import LSTM, Dense, Dropout, MultiHeadAttention, LayerNormalization
import tensorflow as tf
import matplotlib.pyplot as plt
import datetime

# Load and preprocess generation data
gen_data = pd.read_csv('Datasets\Processed Datasets\Plant1_filtered.csv')  # Adjust path
gen_data['DATE_TIME'] = pd.to_datetime(gen_data['DATE_TIME'], format='%d-%m-%Y %H:%M')

# Aggregate AC_POWER across inverters
gen_agg = gen_data.groupby('DATE_TIME')['AC_POWER'].sum().reset_index()
gen_agg.columns = ['DATE_TIME', 'total_AC_POWER']

# Load and preprocess weather data
weather_data = pd.read_csv('Datasets\Processed Datasets\Plant1_Weather_filtered.csv')  # Adjust path
weather_data['DATE_TIME'] = pd.to_datetime(weather_data['DATE_TIME'])

# Merge datasets
data = pd.merge(gen_agg, weather_data[['DATE_TIME', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']], 
                on='DATE_TIME', how='inner')

# Add time features
data['hour'] = data['DATE_TIME'].dt.hour
data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)

# Add Fourier features for daily periodicity (96 time steps = 24 hours * 4 per hour)
def generate_fourier_features(df, period, n_terms):
    df = df.copy()
    df['time_index'] = range(len(df))
    for i in range(1, n_terms + 1):
        df[f'sin_{i}'] = np.sin(2 * np.pi * i * df['time_index'] / period)
        df[f'cos_{i}'] = np.cos(2 * np.pi * i * df['time_index'] / period)
    df = df.drop(columns=['time_index'])
    return df

period = 96  # Daily periodicity for 15-minute intervals
n_terms = 5  # Number of Fourier terms
data = generate_fourier_features(data, period, n_terms)

# Define features and target
features = ['total_AC_POWER', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION', 
            'hour_sin', 'hour_cos'] + [f'sin_{i}' for i in range(1, n_terms + 1)] + [f'cos_{i}' for i in range(1, n_terms + 1)]
target = 'total_AC_POWER'

# Normalize features
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[features])
scaled_df = pd.DataFrame(scaled_data, columns=features, index=data.index)
scaled_df['DATE_TIME'] = data['DATE_TIME']
scaled_df[target] = scaled_data[:, 0]  # Target is first column (total_AC_POWER)

# Define horizons in steps (15-minute intervals)
horizons = {
    '1_hour': 4,    # 1 hour
    '5_hours': 20,  # 5 hours
    '1_day': 96,    # 1 day
    '3_days': 288   # 3 days
}

# Create sequences with multiple target averages
def create_sequences(df, seq_length, target_col, horizons):
    X, y = [], []
    max_horizon = max(horizons.values())
    for i in range(len(df) - seq_length - max_horizon):
        X.append(df[features].iloc[i:i + seq_length].values)
        y_avg = []
        for h in horizons.values():
            if i + seq_length + h < len(df):
                avg = df[target_col].iloc[i + seq_length:i + seq_length + h].mean()
            else:
                avg = np.nan
            y_avg.append(avg)
        y.append(y_avg)
    # Remove rows with NaN targets
    valid_indices = [i for i, y_i in enumerate(y) if not np.isnan(y_i).any()]
    X = [X[i] for i in valid_indices]
    y = [y[i] for i in valid_indices]
    return np.array(X), np.array(y)

seq_length = 24  # 6 hours (24 * 15 min)
X, y = create_sequences(scaled_df, seq_length, target, horizons)

# Split data chronologically
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build hybrid LSTM model with attention for multi-horizon prediction
inputs = tf.keras.Input(shape=(seq_length, len(features)))
inputs_proj = Dense(100)(inputs)  # Project inputs to match LSTM dimension
x = LSTM(100, activation='relu', return_sequences=True)(inputs)
x = Dropout(0.2)(x)
x = MultiHeadAttention(num_heads=4, key_dim=64)(x, x)
x = LayerNormalization()(x + inputs_proj)  # Residual connection
x = LSTM(50, activation='relu')(x)
x = Dropout(0.2)(x)
outputs = Dense(len(horizons))(x)  # Output for each horizon
model = tf.keras.Model(inputs, outputs)
model.compile(optimizer='adam', loss='mse')

# Early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, 
                    callbacks=[early_stopping], verbose=1)

# Predict
y_pred = model.predict(X_test)

# Inverse scale predictions
y_test_rescaled = np.zeros_like(y_test)
y_pred_rescaled = np.zeros_like(y_pred)
for i in range(len(horizons)):
    y_test_rescaled[:, i] = scaler.inverse_transform(np.concatenate([y_test[:, i].reshape(-1, 1), np.zeros((len(y_test), len(features)-1))], axis=1))[:, 0]
    y_pred_rescaled[:, i] = scaler.inverse_transform(np.concatenate([y_pred[:, i].reshape(-1, 1), np.zeros((len(y_pred), len(features)-1))], axis=1))[:, 0]

# Evaluate for each horizon
print("\nEvaluation on test set:")
for i, (label, steps) in enumerate(horizons.items()):
    mae = mean_absolute_error(y_test_rescaled[:, i], y_pred_rescaled[:, i])
    rmse = np.sqrt(mean_squared_error(y_test_rescaled[:, i], y_pred_rescaled[:, i]))
    mape = np.mean(np.abs((y_test_rescaled[:, i] - y_pred_rescaled[:, i]) / (y_test_rescaled[:, i] + 1e-10))) * 100
    print(f'{label} - MAE: {mae:.4f}, RMSE: {rmse:.4f}, MAPE: {mape:.4f}%')

# Inference: Predict future averages based on current input
print("\nEnter current data for prediction:")
current_ac = float(input("Enter current total_AC_POWER (0 to 10000): "))
current_amb_temp = float(input("Enter current AMBIENT_TEMPERATURE (0 to 50): "))
current_mod_temp = float(input("Enter current MODULE_TEMPERATURE (0 to 70): "))
current_irr = float(input("Enter current IRRADIATION (0 to 1.5): "))

# Create a sequence based on the current input
sequence = pd.DataFrame({
    'total_AC_POWER': [current_ac] * seq_length,
    'AMBIENT_TEMPERATURE': [current_amb_temp] * seq_length,
    'MODULE_TEMPERATURE': [current_mod_temp] * seq_length,
    'IRRADIATION': [current_irr] * seq_length,
    'hour_sin': [np.sin(2 * np.pi * datetime.datetime.now().hour / 24)] * seq_length,
    'hour_cos': [np.cos(2 * np.pi * datetime.datetime.now().hour / 24)] * seq_length,
})
sequence['time_index'] = range(len(sequence))
for i in range(1, n_terms + 1):
    sequence[f'sin_{i}'] = np.sin(2 * np.pi * i * sequence['time_index'] / period)
    sequence[f'cos_{i}'] = np.cos(2 * np.pi * i * sequence['time_index'] / period)
sequence = sequence.drop(columns=['time_index'])

# Scale the sequence
sequence_scaled = scaler.transform(sequence[features])

# Predict for all horizons
predictions_scaled = model.predict(sequence_scaled[np.newaxis, :, :])[0]
predictions_rescaled = []
for pred in predictions_scaled:
    pred_array = np.concatenate([[pred], np.zeros(len(features)-1)])
    pred_rescaled = scaler.inverse_transform(pred_array.reshape(1, -1))[0, 0]
    predictions_rescaled.append(pred_rescaled)

# Print predictions for all horizons
print("\nPredictions for all horizons:")
for i, (label, steps) in enumerate(horizons.items()):
    print(f"Predicted average AC_POWER for next {label}: {predictions_rescaled[i]:.2f} kW")

# Plot training history
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Plant 1: Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
